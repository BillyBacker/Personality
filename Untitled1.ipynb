```python
import numba as nb
import numpy as np
from collections import Counter
from numba import jit,cuda, float64
from numpy import genfromtxt,savetxt
import pandas as pd
import math
```

```python
prsonallity = genfromtxt("Cleaned.csv",delimiter=' ')
print(prsonallity.shape)
```

```python
dataset = prsonallity
pinSet = np.random.rand(2,dataset.shape[1])*scalar
resTensor = np.zeros((dataset.shape[0],pinSet.shape[0]))
resVec = np.zeros(dataset.shape[0])
SumArray = np.zeros((dataset.shape[0],pinSet.shape[0]))
dataset_gpu = cuda.to_device(dataset)
pinSet_gpu = cuda.to_device(pinSet)
resTensor_gpu = cuda.to_device(resTensor)
SumArray_gpu = cuda.to_device(SumArray)
resVec_gpu = cuda.to_device(resVec)
```

```python
def genData(k,scalar):
    global dataset, pinSet, resTensor, resVec, SumArray, dataset_gpu, pinSet_gpu, resTensor_gpu, SumArray_gpu, resVec_gpu
    dataset = prsonallity
    pinSet = np.random.rand(k,dataset.shape[1])*scalar
    resTensor = np.zeros((dataset.shape[0],pinSet.shape[0]))
    resVec = np.zeros(dataset.shape[0])
    SumArray = np.zeros((dataset.shape[0],pinSet.shape[0]))
    dataset_gpu = cuda.to_device(dataset)
    pinSet_gpu = cuda.to_device(pinSet)
    resTensor_gpu = cuda.to_device(resTensor)
    SumArray_gpu = cuda.to_device(SumArray)
    resVec_gpu = cuda.to_device(resVec)
def clear():
    global dataset_gpu, pinSet_gpu, resTensor_gpu, SumArray_gpu, resVec_gpu
    dataset_gpu = None
    pinSet_gpu = None
    resTensor_gpu = None
    SumArray_gpu = None
    resVec_gpu = None
```

```python
@jit
def getDigit(num):
    k = 0
    while num/(10**k) >= 1:
        k += 1
    return k
@cuda.jit
def subtractPowSum(dataset, pin, resTensor, e):
    x, y = cuda.grid(2)
    if x < resTensor.shape[0] and y < resTensor.shape[1]:
        for n in range(dataset.shape[1]):
            resTensor[x,y] += (dataset[x,n] - pin[y,n])**e
        resTensor[x,y] = resTensor[x,y]**(1/2)
@cuda.jit
def Min(array, resVec):
    x = cuda.grid(1)
    if x < len(resVec):
        MinVar = 999999
        MinIndex = -1
        for k in range(array.shape[1]):
            if MinVar > array[x,k]:
                MinIndex = k
                MinVar = array[x,k]
        resVec[x] = MinIndex
@cuda.jit
def update(pin, minVec, dataset):
    x,y = cuda.grid(2)
    if x < pin.shape[0] and y < pin.shape[1]:
        pin[x,y]  = 0.0
        n = 0.0
        buffer = 0.0
        for i in range(dataset.shape[0]):
            if x == int(minVec[i]):
                buffer = buffer + dataset[i,y]
                n = n + 1.0
        pin[x,y]  = buffer/n
```

```python
def findGroup():
#     print('Dataset : \n',dataset_gpu.copy_to_host(),end='\n\n')
    
    tpb = (32,32)
    bpgX = int(math.ceil(dataset_gpu.shape[0]/tpb[0]))
    bpgY = int(math.ceil(dataset_gpu.shape[1]/tpb[1]))
    bpg = (bpgX, bpgY)
    subtractPowSum[bpg,tpb](dataset_gpu, pinSet_gpu, resTensor_gpu,2)
#     print('Subpow : \n',resTensor_gpu.copy_to_host())
    bpgX = int(math.ceil(resTensor_gpu.shape[0]/tpb[0]))
    bpgY = int(math.ceil(resTensor_gpu.shape[1]/tpb[1]))
    bpg = (bpgX, bpgY)
    Min[bpg,tpb](resTensor_gpu, resVec_gpu)
    print('Min : \n',Counter(resVec_gpu.copy_to_host()))
    bpgX = int(math.ceil(pinSet.shape[0]/tpb[0]))
    bpgY = int(math.ceil(pinSet.shape[1]/tpb[1]))
    bpg = (bpgX, bpgY)
#     print('Centroids : \n',pinSet_gpu.copy_to_host(),end='\n\n')
    update[bpg,tpb](pinSet_gpu, resVec_gpu, dataset_gpu)
#     print('Updated centroids : \n',pinSet_gpu.copy_to_host(),end='\n\n')
#     return resVec_gpu.copy_to_host().tolist().count(1.0)
    
```

```python
import timeit
epoch = 1
c = (resVec_gpu.copy_to_host().tolist()).count(1.0)
last_c = -1
print(c, last_c)
while c != last_c:
    delta = c - last_c
    print(f"Epochs : {epoch}, Delta : {delta}")
    epoch += 1
    last_c = c
    c = findGroup()
```

```python
specSheet = [{'k':2, 'n':5},
             {'k':3, 'n':5},
             {'k':4, 'n':5},
             {'k':5, 'n':5},
             {'k':6, 'n':5},
             {'k':7, 'n':5},
            {'k':8, 'n':5}]
for s in specSheet:
    for i in range(s['n']):
        name = f"no{i}_k{s['k']}_n1M_range"
        clear()
        genData(s['k'],5)
        epoch = 1
        deltaBool = [False for i in range(s['k'])]
        c = [(resVec_gpu.copy_to_host().tolist()).count(float64(i)) for i in range(s['k'])]
        last_c = [-1 for i in range(s['k'])]
        print(c, last_c)
        while not all(deltaBool):
            c = [(resVec_gpu.copy_to_host().tolist()).count(float64(i)) for i in range(s['k'])]
            deltaBool = [c[i]-last_c[i] == 0 for i in range(s['k'])]
            delta = [last_c[i]-c[i] for i in range(s['k'])]
            print(f"k : {s['k']}, n : {s['n']}, no : {i}, Epochs : {epoch}, Delta : {delta}, range : {max(c)-min(c)}")
            epoch += 1
            last_c = c
            findGroup()
        savetxt(name+f"{max(c)-min(c)}.csv",pinSet_gpu.copy_to_host())
```
